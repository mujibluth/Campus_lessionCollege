### A. Materi (Sample program Fuzzifikasi)

```python
%matplotlib inline

import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl 
from matplotlib import pyplot as plt
```

```python
# x = [0, 1, 2, 3 , .... , 10]
x = np.arange(11)
mfx = fuzz.trimf(x, [0, 5, 10])

print('x = ', x)
print('mfx = ',mfx)

plt.title("Fuzzy Membership Function")
plt.plot(x, mfx, label='normal')
plt.legend(loc="center")
plt.show()
```

```python
# [0, 10]
start = 0
stop = 10 + 0.001
step = 0.25
x = np.arange(start, stop, step)
#print(x)

# Triangular membership function
trimf = fuzz.trimf(x, [0, 5, 10])

# Trapezoidal membership function
trapmf = fuzz.trapmf(x, [0, 2, 8, 10])

# Sigmoid membership function
center = 5.0
width_control = 2.0
sigmf = fuzz.sigmf(x, center, width_control)

# S-function
foot = 1.0
ceiling = 9.0
smf = fuzz.smf(x, foot, ceiling)

# Z-function
zmf = fuzz.zmf(x, foot, ceiling)

# Pi-function
pimf = fuzz.pimf(x, 0.0, 4.0, 5.0, 10.0)

# Gaussian function
mean = 5.0
sigma = 1.25
gaussmf = fuzz.gaussmf(x, mean, sigma)

# Generalized Bell-Shaped function
width = 2.0
slope = 4.0
center = 5.0
gbellmf = fuzz.gbellmf(x, width, slope, center)

fig_scale = 1.5
plt.figure(figsize=(6.4 * fig_scale, 4.8 * fig_scale))

# 3 rows, 1 col, index from 1
plt.subplot(311)
plt.title("Fuzzy Membership Function")
plt.plot(x, trimf, label="Triangle")
plt.plot(x, trapmf, label="Trapezoidal")
plt.legend(loc="upper right")

plt.subplot(312)
plt.plot(x, sigmf, label="Sigmoid")
plt.plot(x, smf, label="S-function")
plt.plot(x, zmf, label="Z-function")
plt.plot(x, pimf, label="Pi-function")
plt.legend(loc="upper right")

plt.subplot(313)
plt.plot(x, gaussmf, label="Gaussan")
plt.plot(x, gbellmf, label="Bell-Shaped")
plt.legend(loc="upper right")

plt.show()
```

```python
## Fuzzification Example

x = np.arange(91)
xmf_e = fuzz.trimf(x, [0,0,45])
xmf_n = fuzz.trimf(x, [0,45,90])
xmf_h = fuzz.trimf(x, [45,90,90])

xmf = [xmf_e, xmf_n, xmf_h]
label = ['dingin', 'normal', 'panas']
for xmfs, labels in zip(xmf, label):
    plt.plot(x, xmfs, label=labels)
plt.legend(loc="center")
plt.show()

z_val = []
for i in range(len(xmf)):
    #z.append(xmf[i])
    z = fuzz.interp_membership(x, xmf[i], 25)
    z_val.append(z)
    
print(z_val)
```

### B. Contoh Fuzzifikasi dengan permasalahan nyata

```python
%matplotlib inline

import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl 
from matplotlib import pyplot as plt
```

```python
## Modify fuzzification using 5 category

x = np.arange(100)
xmf_j = fuzz.trimf(x, [0,20,30])
xmf_k = fuzz.trimf(x, [20,30,40])
xmf_l = fuzz.trimf(x, [30,40,50])
xmf_m = fuzz.trimf(x, [40,50,60])
xmf_n = fuzz.trimf(x, [50,60,80])

xmf = [xmf_j, xmf_k, xmf_l, xmf_m, xmf_n]
label = ['sangat pelan', 'pelan', 'normal', 'cepat', 'sangat cepat']
for xmfs, labels in zip(xmf, label):
    plt.plot(x, xmfs, label=labels)
plt.legend(loc="upper right")
plt.show()

z_val = []
for i in range(len(xmf)):
    #z.append(xmf[i])
    z = fuzz.interp_membership(x, xmf[i], 25)
    z_val.append(z)
    
print(z_val)
```

Tugas 1 : Membuat fuzzifikasi dengan 5 kategory.
Pada kali ini, saya menggunakan kategory kecepatan untuk digunakan sebagai pembanding nilai diagramnya. Nilai maksimum yang saya tampilkan adalah 80, tetapi saya menggunakan angka 100 untuk nilai maksimum dibuatkanya diagram tersebut. Hal ini saya lakukan karena saya ingin menempatkan legenda di pojok kanan atas (fungsi 'plt.legend') dengan backgound kosong/ tidak menutupi garis grafik yang saya buat

### C. Penggunaan 'automf pada skfuzzy' untuk point "B"

```python
## Fuzzification using 'automf'

x = np.arange(100)
xmf = ctrl.Antecedent(x, 'xmf')
names = ['sangat pelan', 'pelan', 'normal', 'cepat', 'sangat cepat']
xmf.automf(names=names)

xmf.view()
```

Tugas 2 : Membuat fuzzifikasi menggunakan 'auotomf'. Parameter untuk menjalankan 'automf' adlah menggunakan names. sebelumnya kita membuat 'Antacedent' dengan parameter library numpy lalu memanggil fungsi xmf. Dimana fungsi xmf mendeklarasikan nama variabel yang berisi nilai kategori yang kita buat.

Lalu untuk menggunakan 'automf' kita bisa menggunakan fungsi xmf dengan parameter 'names' yang isinya kategory yang akan kita presentasikan/ tampilkan. Salah satu keuntungan kita menggunakan 'automf' adalah kita juga bisa membuat garis diagram lebih dari satu, bisa mendeklarasikan banyak variabel dalam satu diagram sekaligus. Keuntungan yang lain adalah kita juga bisa men-generate nilai secara otomatis. Hal ini tentu sangat membantu developer dalam membuat programnya karena bisa menyingkat program

### D. Fungsi keanggotaan pada 3 titik pengukuran menggunakan 'interp_membership'

```python
## Fuzzification using 'interp_membership'

x = np.arange(81)
xmf_e = fuzz.trimf(x, [0,0,40])
xmf_n = fuzz.trimf(x, [0,40,80])
xmf_h = fuzz.trimf(x, [40,80,80])

xmf = [xmf_e, xmf_n, xmf_h]
label = ['pelan', 'normal', 'cepat']
for xmfs, labels in zip(xmf, label):
    plt.plot(x, xmfs, label=labels)
plt.legend(loc="center")
plt.show()

z_val = []
for i in range(len(xmf)):
    #z.append(xmf[i])
    z = fuzz.interp_membership(x, xmf[i], 25)
    z_val.append(z)
    
print(z_val)
```

Perbandingan menggunakan perhitungan manual :

1) pelan  [ 20 ]    = (40-20)/ (40-0)  = 0,3
2) normal [ 30 ]    = (30-0)/ (40-0)   = 0,75
3) cepat  [ 60 ]    = (60-40)/ (80-40) = 0,905

Jika ditarik garis koordinat/ dibuat diagram maka sesuai dengan diagram yang sudah dibuat. Jadi, perhitungan manual ataupun menggunakan program seperti diatas, maka hasilnya sama saja
